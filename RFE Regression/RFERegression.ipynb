```python
import pandas as pd
```

```python
dataset1=pd.read_csv("prep.csv",index_col=None)
df2=dataset1
df2 = pd.get_dummies(df2, drop_first=True)

indep_X=df2.drop('classification_yes', axis=1)
dep_Y=df2['classification_yes']
```

```python
from RFERegression import rfeRegression
rfelist=rfeRegression.rfeFeature(indep_X,dep_Y,5)       
acclin=[]
accsvml=[]
accsvmnl=[]
accdes=[]
accrf=[]
```

```python
for i in rfelist:   
    X_train, X_test, y_train, y_test=rfeRegression.split_scalar(i,dep_Y)  
    r2_lin=rfeRegression.Linear(X_train,y_train,X_test,y_test)
    acclin.append(r2_lin)
    
    r2_sl=rfeRegression.svm_linear(X_train,y_train,X_test,y_test)    
    accsvml.append(r2_sl)
    
    r2_NL=rfeRegression.svm_NL(X_train,y_train,X_test,y_test)
    accsvmnl.append(r2_NL)
    
    r2_d=rfeRegression.Decision(X_train,y_train,X_test,y_test)
    accdes.append(r2_d)
    
    r2_r=rfeRegression.random(X_train,y_train,X_test,y_test)
    accrf.append(r2_r)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
result=rfeRegression.rfe_regression(acclin,accsvml,accdes,accrf)
```
